```
# getting top 250 movies 
import numpy as np
import pandas as pd
mtype = 'S4,S100, S400, S4'
d= np.empty((40,4), dtype=mtype)
search = ia.get_top250_movies() 
 
# printing only first 10 movies title 
index =0
for i in search[:10]: 
    print(i.movieID ) 
    print(i['title'] ) 
   # theMatrix = ia.get_movie(search[i].movieID)
    d[index][0] = i.movieID
    d[index][1] = i['title'].encode('ascii', 'ignore')
    ia.update(i,['reviews'])
    for review in i['reviews'][:2]:
      print(review['content'])
      print(review['rating'])
      d[index][2] = review['content'].encode('ascii', 'ignore')
      d[index][3] = review['rating']
      index = index+1
search = ia.get_bottom100_movies() 
  
# printing only first 10 movies title 
for i in search[:10]: 
    print(i.movieID ) 
    print(i['title'] ) 
   # theMatrix = ia.get_movie(search[i].movieID)
    d[index][0] = i.movieID
    d[index][1] = i['title'].encode('ascii', 'ignore').decode()
    ia.update(i,['reviews'])
    for review in i['reviews'][:2]:
      print(review['content'])
      print(review['rating'])
      d[index][2] = review['content'].encode('ascii', 'ignore').decode()
      d[index][3] = review['rating']
      index = index+1
print(d)
df2 = pd.DataFrame(d,columns=['a','as', 'b', 'c'])
print(df2)
```
